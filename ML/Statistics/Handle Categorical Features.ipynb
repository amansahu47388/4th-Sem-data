```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("mall_customers.csv")
data
```

## One Hot Encoding with Sklearn

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_h = OneHotEncoder(sparse=False)
```

```python
rd = one_h.fit_transform(data[['Genre']])
rd
```

```python
data2=pd.DataFrame(rd,columns=['Genere_female','Genere_male'])
data2
```

```python
final_data2 = pd.concat([data,data2],axis=1)
final_data2
```

```python
final_data3 = final_data2.drop('Genre',axis=1)
final_data3
```

## Label Encoding

```python
data4 = pd.read_csv('name.csv')
data4
```

```python
data5=data4[['size']]
data5
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(data5)
```

```python
data5['size_update']=le.fit_transform(data5['size'])
data5
```

```python
data4['size'] = data5['size_update']
data4
```

```python
data4['area_type'].value_counts()
```

```python
data6=data4[['area_type']]
data6
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit_transform(data6)
```

```python
data6['area_type_new']=le.fit_transform(data6['area_type'])
data6
```

```python
data6.drop('area_update',axis=1)
```

```python
data4['area_type'] = data6['area_type_new']
data4
```

```python
data6=data4[['availability']]
data6
```

```python
data6['availability_new']=le.fit_transform(data6['availability'])
data6
```

```python
data4['availability'] = data6['availability_new']
data4
```
