## IMPLEMENT DATA VISULIZATION USING MATPLOTLIB & SEABORN


```python
import matplotlib.pyplot as plt
```

```python
import  numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

## scatter plot

```python
plt.scatter(x,y,c='b')
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Graph in 2D')
```

## LINE CHART

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'ro--', linewidth=2, markersize=12)
plt.savefig('test.png')
```

## SUBPLOTS

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'yo')
plt.subplot(2,2,4)
plt.plot(x,y,'bo--')
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.subplot(3,3,1)
plt.plot(a,b,'r--')
plt.subplot(3,3,2)
plt.plot(a,b,'g*--')
plt.subplot(3,3,3)
plt.plot(a,b,'yo')
plt.subplot(3,3,4)
plt.plot(a,b,'bo--')
plt.subplot(3,3,5)
plt.plot(a,b,'r--')
plt.subplot(3,3,6)
plt.plot(a,b,'g*--')
```

## BAR PLOT

```python
x=[2,8,10]
y=[11,16,9]
x1=[2,6,3]
y1=[4,9,36]
plt.bar(x,y)
```

```python
x=[2,8,10]
y=[11,16,9]
x1=[2,6,3]
y1=[4,9,36]
plt.bar(x,y)
plt.bar(x1,y1,color='g')
```

## HISTOGRAM PLOT

```python
a= np.array([22,65,94,53,5,36,32,78,22,63,96,14,25,85,12])
plt.hist(a)
```

## pie chart plot

```python
labels='python','c++','ML','Java'
sizes=[263,241,456,120]
colors=['gold','yellowgreen','red','lightcoral']
explode=(0.2,0,0,0)
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.2f%%',explode=explode)
```

```python
labels='M3','DBMS','SEAM','DATA SCIENCE','OS'
sizes=[5,19,22,12,13]
colors=['gold','yellowgreen','red','lightcoral','blue']
explode=(0.2,0,0,0,0)
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.2f%%',explode=explode)
```

# SEABORN LIBRARY

```python
import seaborn as sns
```

```python
df5= sns.load_dataset("tips")
df5
```

```python
df5 = df5.drop(columns=['sex','smoker','day','time'])
```

```python
df5.corr()
```

```python
sns.heatmap(df5.corr())
```

## JOINT PLOT

```python
sns.jointplot(x='tip',y='total_bill',data=df5)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df5, kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df5, kind='reg')
```

## PAIR PLOT

```python
df6= sns.load_dataset("tips")
df6
```

```python
sns.pairplot(df6,hue='sex')
```

## DIST PLOT

```python
sns.distplot(df5['total_bill'])
```
